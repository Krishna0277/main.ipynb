```python
# Palindrome Checker
def palindrome(a):
    m = a
    rev = 0
    while m > 0:
        z = m % 10
        rev = rev*10 + z
        m = m//10
    if rev == a:
        return f"Your Number {a} is a Palindromic Number"
    else:
        return f"Your Number {a} is not a Palindromic Number"


print(palindrome(121))

```
